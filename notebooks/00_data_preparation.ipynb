## **Обработка данных**

Для начала из предоставленных данных необходимо получить два датасета:
- набор на котором будет обучаться модель;
- набор с пользователями, для которых необходимо предсказать пол.

```python
train_merged_df = train_df.merge(geo_info_df, on='geo_id', how='inner')
train_merged_df = train_merged_df.merge(referer_vectors_df, on='referer', how='inner')
train_merged_df = train_merged_df.merge(train_labels_df, on='user_id', how='inner')
train_merged_df = train_merged_df.drop(['geo_id', 'referer'], axis=1)
```

```python
test_merged_df = test_df.merge(geo_info_df, on='geo_id', how='inner')
test_merged_df = test_merged_df.merge(referer_vectors_df, on='referer', how='inner')
test_merged_df = test_merged_df.merge(test_users_df, on='user_id', how='inner')
test_merged_df = test_merged_df.drop(['geo_id', 'referer'], axis=1)
```

Удалим дубликаты из полученных датасетов.

```python
train_merged_df = train_merged_df.drop_duplicates()
test_merged_df = test_merged_df.drop_duplicates()
```

Посмотрим на кол-во пропущенных значений.

```python
train_missing = train_merged_df.isna().sum().sort_values(ascending=False)

pd.DataFrame({"Missing": train_missing, "Missing (%)": (train_missing / len(train_merged_df) * 100)})
```

```python
test_missing = test_merged_df.isna().sum().sort_values(ascending=False)

pd.DataFrame({"Missing": test_missing, "Missing (%)": (test_missing / len(test_merged_df) * 100)})
```

Распарсим столбец **user_agent**, предварительно удалив строку с пропущенным значением в данном столбце.

Пропуски в **region_id** обработаем на следующем шаге.

```python
train_merged_df = train_merged_df.dropna(subset=['user_agent'])
```

```python
import ast

train_agent_df = train_merged_df['user_agent'].apply(ast.literal_eval)
train_agent_df = train_agent_df.apply(pd.Series)
train_merged_df = train_merged_df.drop(columns=['user_agent']).join(train_agent_df)

test_agent_df = test_merged_df['user_agent'].apply(ast.literal_eval)
test_agent_df = test_agent_df.apply(pd.Series)
test_merged_df = test_merged_df.drop(columns=['user_agent']).join(test_agent_df)
```

## **Скачивание полученных датасетов**

```python
order = [
    'user_id',
    'component0', 'component1', 'component2', 'component3', 'component4', 'component5', 'component6', 'component7', 'component8', 'component9',
    'request_ts', 'country_id', 'region_id', 'timezone_id', 'browser', 'browser_version', 'os', 'os_version',
    'target'
]

train_merged_df = train_merged_df[order]
test_merged_df = test_merged_df[order[:-1]]
```
